##программа

```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
def get_input(commands):
    while True :
      # for c in commands : print(f'{c} - {commands[c]}')
      com = input('\nВведите команду: \n')
      print('')
      if com in commands : return com
      else : 
        print('!Неизвестная команда!')
        continue
```

```python
# def q(): #команда завершить программу
#     a=0
```

```python
def p(): #команда узнать владельца по номеру
    name = None
    num = input('Введите номер документа: \n')
    name = [doc['name'] for doc in documents if doc['number'] == num] 
    '''
    перебрать все записи о документах и составить список тех [имен] владельцев,
    для которых [номер] совпадает с введенным

    предполагается только одно совпадение, поэтому [0]
    '''
    if name == [] : print('Документ не найден в базе')
    else : print(f'Владелец документа: {name[0]}')
```

```python
def get_shelf(num) : #полка документа по номеру
    dir = [ dir for dir in directories if num in directories[dir] ]
    '''
    перебрать все записи о полках и составить список имен тех,
    "в" которых содержится введенный номер

    предполагается только одно совпадение, поэтому [0]
    '''
    if dir == [] : return 'error'
    else : return dir[0]       
```

```python
def shelf_list_print () : #вывести список полок через запятую
    shelf_list = ', '.join(directories.keys())
    print(f"Текущий перечень полок: {shelf_list}.")
```

```python
def s(): #команда полка документа по номеру
    dir = None
    num = input('Введите номер документа: \n')
    s = get_shelf(num)
    if s == 'error' : print('Документ не найден в базе')
    else : print(f'Документ хранится на полке: {s}')    
    
```

```python
def l(): #команда информация по всем документам
    for doc in documents :
      print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {get_shelf(doc['number'])} ")
      
```

```python
def ads(): #команда добавить полку
    s = input('Введите номер полки: \n')
    if s in directories : print('Такая полка уже существует. ', end='') 
    else :
       directories[s] = []
       print('Полка добавлена. ', end='')

    shelf_list_print()
    # shelf_list = ', '.join(directories.keys())
    # print(f"Текущий перечень полок: {shelf_list}.")
    # for s in directories : print(s, end=', ') 
    
    print('')
```

```python
def ds(): #удалить полку
    s = input('Введите номер полки для удаления: \n')
    if s in directories : 
       if directories[s] == [] : 
          del directories[s] 
          print('Полка удалена. ', end='')
          shelf_list_print()
       else :
          print('На полке есть документs, удалите их перед удалением полки. ',end='')
          shelf_list_print()
    else :
       print('Такой полки не существует. ', end='')
       shelf_list_print()
```

```python
def add_doc(num,t,n,s) : #добавление документа в базу
    documents.append({'type' : t, 
                      'number':num, 
                      'name' : n
                      })
    directories[s].append(num)
```

```python
def ad() : #команда добавление документа
    num = input('Введите номер документа: \n')
    if get_shelf(num) != 'error' : 
       print('Документ с таким номером уже присутствует в базе. Удалите его командой "d".')
      #  return
    else :
      t = input('Введите тип документа: \n')
      n = input('Введите владельца документа: \n')
      s = input('Введите полку для хранения: \n')
      if s not in directories :
         print('Такой полки не существует. Добавьте полку командой as.')
        #  return
      else :
        print('Документ добавлен. ', end = '')
        add_doc(num,t,n,s)

    print('Текущий список документов:')
    l()
```

```python
def d() : #команда удаление документа
    num = input('Введите номер документа: \n')
    if get_shelf(num) == 'error' : print('Документ не найден в базе.')
    else :
       [documents.remove(doc) for doc in documents if doc['number'] == num]    
       print('Документ удален.')
    
    print('Текущий список документов:')
    l()
```

```python
def m() : #команда перемещение документа
    num = input('Введите номер документа: \n')
    s = get_shelf(num)
    if s == 'error' : print('Документ не найден в базе.')
    else : 
       s_new = input('Введите номер полки: \n')
       if s_new not in directories :
          print('Такой полки не существует. Добавьте полку командой as.')
       else :
         directories[s].remove(num)
         directories[s_new].append(num)
         print('Документ перемещен.')    
         
    print('Текущий список документов:')
    l()   
    
```

```python
def main():
  commands = {'q' : 'завершить программу',
              'p' : 'узнать владельца по номеру',
              's' : 'полка документа по номеру',
              'l' : 'информация по всем документам',
              'ads':'добавить полку',
              'ds': 'удалить полку',
              'ad': 'добавить документ',
              'd' : 'удалить документ',
              'm' : 'переместить документ'
              }



  while True:
    # print('')
    com = get_input(commands)
    # print(com)

    if com == 'q' : break
    elif com=='p' : p()
    elif com=='s' : s()
    elif com=='l' : l()
    elif com=='ads':ads()
    elif com=='ds': ds()
    elif com=='ad': ad()
    elif com=='d' : d()
    elif com=='m' : m()
    else : print('!Ошибка!')
```

```python
main()
```
