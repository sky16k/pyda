Напишем программу для проверки логина и пароля

1) логин и пароль пользователя хранится в переменной

2) будем давать пользователю k попыток на ввод логина и пароля

3) если после трех попыток вверного ввода не было, то заблокируем вход - вывести на экран сообщение "Вход заблокирован"

4) если пароль и логин были введены верно, то вывести на экран сообщение "Добро пожаловать!"

```python
login_correct = 'admin'
pswd_correct = 'admin'
k = 3
for i in range(0, k):
    login = input('Ввелите логин')
    pswd = input('Введите пароль')
    if login != login_correct or pswd != pswd_correct:
        k -= 1
        print(f'Осталось {k} попытки')
    else:
        print('Добро пожаловать!')
        break
else:
    print('Вход заблокирован')
```

Посчитаем сколько раз каждое слово встречалось в тексте и выведем информацию в виде:  
`"слово": "количество вхождений" `


```python
some_oxy_text = '''
Весь мой рэп, если коротко, про то, что
Уж который год который город под подошвой
В гору, когда прет. Потом под гору, когда тошно
Я не то, что Гулливер, но все же город под подошвой
Город под подошвой, город под подошвой
Светофоры, госпошлины, сборы и таможни
Я не знаю, вброд или на дно эта дорожка
Ты живешь под каблуком, у меня - город под подошвой
'''
```

```python
res = {}

splitted_text = some_oxy_text.lower().replace('-', '').split()
for word in splitted_text:
    word = word.strip(',.!-')
    # print(word)
    res.setdefault(word, 0)
    res[word] += 1

for word, count in res.items():
    print(f'{word}: {count}')
```

```python
list_ = []
for row in some_oxy_text.lower().split():
    list_.append(row.strip(',..,?-!'))
    res = set(list_)
for el in res:
  print(f'слово:{el} количество вхождений:  {list_.count(el)}')
```

```python
clean_text = [word.strip(',..,?!-') for word in some_oxy_text.lower().replace('-', '').split()]
for word in set(clean_text):
    print(f'слово:  {word} количество вхождений:  {clean_text.count(word)}')
```

Посмотрим, как знание сложных структур может помочь в задачах, приближенных к реальным

```python
import requests
import time
# функция pprint позволяет более структурированно выводить сложные структуры
from pprint import pprint
token = ''
```

Поищем сообщества в социальной сети

```python
params = {
    'q': 'python',
    'access_token': token,
    'v':'5.131',
    'count': 300
}
req = requests.get('https://api.vk.com/method/groups.search', params).json()
```

```python
print(req)
```

```python
pprint(req)
```

```python
target_groups = req['response']['items']
target_groups
```

```python
# преобразуем список всех id в строку (в таком виде принимает данные параметр fields)
target_group_ids = ','.join([str(group['id']) for group in target_groups])
target_group_ids
```

```python
params = {
    'access_token': token,
    'v':'5.131',
    'group_ids': target_group_ids,
    'fields':  'members_count,activity,description'

}
req = requests.get('https://api.vk.com/method/groups.getById', params)

pprint(req.json())
```

Посчитаем количество лайков в сообщениях

```python
params = {
    'access_token': token,
    'v':'5.131',
    'q': 'Нетология',
    'start_from': '',
    'count': 200,
    'extended': 1
}   
res = []
while True:
    result = requests.get('https://api.vk.com/method/newsfeed.search', params)
    time.sleep(0.33)
    res += result.json()['response']['items']
    if 'next_from' in result.json()['response'].keys():
        params['start_from'] = result.json()['response']['next_from']
    else:
        break
```

```python
print(type(res))
print(len(res))
```

```python
pprint(res[0])
```

```python
likes_count = 0
reposts_count = 0
for post in res:
    likes_count += post['likes']['count']
    reposts_count += post['reposts']['count']

print(likes_count)
print(reposts_count)
```
