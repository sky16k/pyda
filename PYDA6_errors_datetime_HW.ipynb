###программа

```python
from datetime import datetime
```

```python
stream = ['2018-04-02', '2018-02-29', '2018-19-02']
```

```python
def date_verif(st) :
    try :
      date = datetime.strptime(st, '%Y-%m-%d')
      return (type(date) == datetime)
    except :
      return ('False')
```

```python
for line in stream :
    print(f'{line} - {date_verif(line)}')
```

```python
def date_range() :  #(sd_str, ed_str) # строки начальной и конечной дат 
    sd_str = input('Введите начальную дату (в формате YYYY-MM-DD: ')
    try :
      sd = datetime.strptime(sd_str, '%Y-%m-%d') #начальная дата
    except ValueError: return []

    ed_str = input('Введите конечную дату (в формате YYYY-MM-DD: ')
    try :
      ed = datetime.strptime(ed_str, '%Y-%m-%d') #конечная дата
    except ValueError: return []

    if sd > ed : return []
    else:
      d_range = [sd]
      d = sd
      while d < ed :
        d += timedelta(days=1)
        d_range.append(d)
      return d_range
```

>Что значит ошибка list index out of range?

обращение к списку по индексу вне имеющегося диапазона.
например: список длиной 5, а обращение к 10-му элементу или "-6"-му

>Почему при первом запуске функция работает корректно, а при втором - нет?

функция использует положительную индексацию, но смещение берет с конца (-2), опираясь, при этом на неизменяемую глобальную переменную (деф_юз_кнт)
при первом запуске, даже после удаления последнего элемента, получаемый индекс всё ещё попадает в диапазон, указывая на последний элемент итогового списка (равный, кстати, 1) 
во второй раз удаляется ещё один элемент, а индекс не меняется (поскольку высчитывается на основе неизменённой переменной) и указывает на только что удалённый элемент. Т.о. индекс оказывается вне диапазона, что и вызывает ошибку.
